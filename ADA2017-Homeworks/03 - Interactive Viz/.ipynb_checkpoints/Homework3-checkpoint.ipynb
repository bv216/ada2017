Exploring interactive visualization from two different datasets, the statistics offices of the European commission and the Swiss Confederation.

## Task 1 - European unemployment rates

For this task we need the unemployment rates from [Eurostat](http://ec.europa.eu). We go into Eurostat and search 'Unemnployment rate' and sort by 'Publication date'. Best result can be found [here](http://ec.europa.eu/eurostat/web/products-datasets/-/tipsun20), which is the Unemployment rate - annual data updated just a few days ago.
    

```python
    #Save the eu unemployment rates into euData and show some info
euData = pd.read_excel('Eurostat_Unemployment_rate.xls', header=2) 
euData.head()
```

```python
    #Cleaning euData, quit unnecessary years and rows without relevant info.
euData = euData.iloc[0:28]
euData = euData[['geo\\time','2016']]
euData.columns = ['Country','Unemployment']
#euData = euData.set_index('Country')
euData.head()
```

Now that we have the data, we are going to plot it on a Choropleth map. 

We start by creating the map and adding some layers with the data as a choropleth map.

```python
    #Overlay the country limits.
state_geo_path= 'europe.topojson.json'
geo_json_data = json.load(open(state_geo_path))

    #Create the map
eu_map = folium.Map(location=[53, 8.33],
                   tiles='cartodbpositron', zoom_start=3.5)

eu_topo_path = os.path.join('', 'europe.topojson.json')

eu_map.choropleth(geo_data=geo_json_data, 
                  topojson='objects.europe',
                  data=euData,
                  columns=['Country', 'Unemployment'],
                  key_on='feature.properties.NAME',
                  fill_color='OrRd', fill_opacity=0.5, line_opacity=0.5)
    
eu_map.save('UE_unemployment.html')

    #Showing the map
eu_map
```

What we can appreciate from this map is that most of the mediterranian countries are the darkness ones while the north of Europe is closer to white.

Let's compare them also by using a bar chart.


```python
    #barchart for Eu unemployment % in 2016
sns.set_style('darkgrid')
fig, ax = plot.subplots(figsize = (15,8))
ax.set_title('Unemployment in Europe in 2016', fontsize=15, fontweight='bold')
sns.barplot(x='Country', y='Unemployment', data=euData.sort_values(['Unemployment'],ascending=0), saturation=0.7, errcolor='.7')
plot.xticks(rotation=90)
plot.show()
```

From this graph, as It was mentioned before, Greece stands out above the rest of countries, together with Spain, with an unemployment close or even higher to the 20%.
On the other side, there are two countries that are Germany and Czech Republic with values around 4%. Followed very closed by many others like Hungary or United Kingdom.


## Task 2 - Switzertland unemployment rates

In the following cell, we create the path and read the *topojson* file that will enable us to have each canton's geometry on the map.

```python
state_geo_path= './topojson/ch-cantons.topojson.json'
geo_json_data = json.load(open(state_geo_path))
```

Here is the first map created, centered on Switzerland.

```python
m_ch = folium.Map(location=[46.8, 8.33],
                   tiles='Mapbox Bright', zoom_start=7)
m_ch
```

The file *ch-cantons.topojson.json* is added as a layer on top of the initial map.

```python
ch_topo_path = os.path.join('topojson', 'ch-cantons.topojson.json')

folium.TopoJson(
    open(ch_topo_path),
    'objects.cantons',
    name='topojson'
).add_to(m_ch)

m_ch
```

Now, we read the data downloaded from the website of the Swiss Statistics Office (*amstat*).

From the extensive dataset, we only selected 4 variables describing the overall unemployment phenomena in Switzerland. We thus have for each canton (26) these 4 values.

The *Unemployment Rate* corresponds to the ratio of the unemployed population over the overall population [%].

The *Unemployed* are the people in the age of working without a job. They might not be looking for one.

Thus the *Jobseeker* category is the population looking for a job including the people having a first job.

Finally, the *Employed Jobseeker* represents this last category of people.

```python
unemployment=pd.read_csv('chomeurs.csv',encoding='latin-1',thousands="'")
unemployment.columns=['Canton','Unemployment Rate','Unemployed','Jobseeker','Employed Jobseeker']
unemployment.head()
```

Now, we get the id of canton in the *json* file. Luckily, even if the name of each canton may vary in the *json* and *unemployment* files (*Zurich* and *Zürich* for example), the order is the same. 

This allows us to get the *ID* very simply.

```python
id_canton=[canton['id'] for canton in geo_json_data['objects']['cantons']['geometries']]
unemployment['ID']=id_canton

unemployment.head()
```

Finally, we use the *Unemployment* table and the *choropleth()* function to map the spatial distribution of unemployment indicators.

Each time we recreate the map with the *folium.Map* function to get rid of the overlayers. 

First, the different values of *Unemployment Rate* for each canton.

```python
ch_map = folium.Map(location=[46.8, 8.33],
                   tiles='Mapbox Bright', zoom_start=7)

ch_map.choropleth(geo_data=geo_json_data, 
                  topojson='objects.cantons',
                  data=unemployment,
                  columns=['ID', 'Unemployment Rate'],
                  key_on='feature.id',
                  fill_color='OrRd', fill_opacity=0.7, line_opacity=0.2,
                  legend_name='Unemployed')
m_ch.save('CH_unemployment.html')

ch_map
```

It seems like the rural cantons tend to have low *unemployment rate*.

Now we compute the *Jobseeker Rate* to map its spatial distribution.

```python
unemployment['Jobseeker Rate'] = round(unemployment['Jobseeker']/(unemployment['Unemployed']/unemployment['Unemployment Rate']),2)
unemployment.head()
```

Unsurprisingly, the *Jobseeker Rate* is always bigger than the *Unemployment Rate*.

```python
ch_map = folium.Map(location=[46.8, 8.33],
                   tiles='Mapbox Bright', zoom_start=7)

ch_map.choropleth(geo_data=geo_json_data, 
                  topojson='objects.cantons',
                  data=unemployment,
                  columns=['ID', 'Jobseeker Rate'],
                  key_on='feature.id',
                  fill_color='OrRd', fill_opacity=0.7, line_opacity=0.2,
                  legend_name='Jobseeker Rate')

m_ch.save('CH_jobseeker.html')

ch_map
```

As it can be seen the french part and the italian part of Switzerland have a higher rate of people looking for a job. There is still the same trend than for unemployement, in the sense that rural parts have smaller portion of people seeking for a job, while in the cantons including a main city as it can be seen in Geneva, Vaud, Basel (a little bit less though) more people are jobseeker. That is all the big cities at the exception of Zurich and maybe Berne.

Let's compute the part of unemployed seeking a job with the *Unemployed Jobseeker Rate*.

```python
unemployment['Unemployed Jobseeker Rate'] = round((1-unemployment['Employed Jobseeker']/unemployment['Jobseeker'])*100,2)
unemployment.head()
```

```python
ch_map = folium.Map(location=[46.8, 8.33],
                   tiles='Mapbox Bright', zoom_start=7)

ch_map.choropleth(geo_data=geo_json_data, 
                  topojson='objects.cantons',
                  data=unemployment,
                  columns=['ID', 'Unemployed Jobseeker Rate'],
                  key_on='feature.id',
                  fill_color='OrRd', fill_opacity=0.7, line_opacity=0.2,
                  legend_name='Unemployed Jobseeker Rate')

m_ch.save('CH_unemployed_jobseeker_ratio.html')

ch_map
```

It seems like the rural jobseekers tend to have already a job. They might be seeking to change job, but it is more likely that they are looking for a second job and thus have a higher 

### Comparing Europe and Switzerland
Now we are going to compare Europe and Switzerland.
For this, we are plotting some relevant information like the quartiles.

Plot for Europe, in green.

```python
plot.subplot(2,1,1)
sns.boxplot(data=euData[['Unemployment']],orient='h',palette="Set2",whis=1.5)
plot.show()
```

Plot for Switzerland, in red.

```python
plot.subplot(2,1,2)
sns.boxplot(data=unemployment[['Unemployment Rate']],orient='h',palette="Set1",whis=1.5)
plot.show()
```

What we see is that almost every cantons of Switzerland are below the great amount of the values from Europe, so that almost all of Switzerland is below Europe in terms of unemployment.
